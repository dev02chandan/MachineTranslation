```python
from transformers import MarianMTModel, MarianTokenizer

def translate_text(sentences, source_lang, target_lang):
    # Construct model name based on source and target languages
    model_name = f'Helsinki-NLP/opus-mt-{source_lang}-{target_lang}'

    # Load tokenizer and model from Hugging Face's Transformers
    tokenizer = MarianTokenizer.from_pretrained(model_name)
    model = MarianMTModel.from_pretrained(model_name)

    # Function to translate a single sentence
    def translate_sentence(sentence, tokenizer, model):
        # Tokenize the text
        model_inputs = tokenizer(sentence, return_tensors="pt", padding=True, truncation=True)

        # Translate
        translation = model.generate(**model_inputs)

        # Decode the translation
        translated_text = tokenizer.decode(translation[0], skip_special_tokens=True)
        return translated_text

    # Translate all sentences in the list
    translated_sentences = [translate_sentence(sentence, tokenizer, model) for sentence in sentences]

    return translated_sentences
```

```python
source_language = "fr"
target_language = "en"

# Translate the sentences
translated_sentences = translate_text(
    french_sentences, source_language, target_language
    )

# Output the translated sentences
for i, translated in enumerate(translated_sentences, 1):
    print(f"{i}: {translated}")
```
